### Using broadcasting create an output matrix based on the following two arrays:

    array_1 = np.array([1,2,3,4,5], dtype=np.int8)
    array_2 = np.array([1,2,3], dtype=np.int8)

Expected output:

    [[ 1  2  3]
     [ 2  4  6]
     [ 3  6  9]
     [ 4  8 12]
     [ 5 10 15]]

```python
array_1 = np.array([1,2,3,4,5], dtype=np.int8).reshape(-1, 1)
print(array_1)
array_2 = np.array([1,2,3], dtype=np.int8)
print(array_2)
matrix = array_1 * array_2

print(matrix)
```
