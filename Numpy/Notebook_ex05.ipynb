```python
import numpy as np
```

### Generate an array with integers from 1 to 50

```python
array1 = np.arange(1, 51)
print(array1)
```

### Generate an array with integers from 51 to 100

```python
array2 = np.arange(51, 101)
print(array2)
```

### Using np.concatenate, concatenate the two arrays into: array([1,...,100])

```python
array = np.concatenate((array1, array2))
print(array)
```
